```python
import random
number_of_points = 10**8
number_inside_circle = 0
random.seed()
for point in range(number_of_points):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    if x**2 + y**2 < 1:
        number_inside_circle += 1

pi_approx = 4.0 * number_inside_circle / number_of_points
print(f"With {number_of_points} of points our approximation of pi is {pi_approx}")
```

```python
import numpy as np 
import matplotlib.pyplot as plt

number_of_points = 10**3
x = np.random.uniform(-1, 1, number_of_points)
y = np.random.uniform(-1, 1, number_of_points)

# compute pi
pi_approx = 4.0 * np.sum(x**2 + y**2 <= 1) / number_of_points

#Sactter plot
fig = plt.figure(figsize=(8, 6), dpi=600)
plt.scatter(x, y, alpha=0.5, color='red')

circle = plt.Circle((0,0), 1, color='blue', alpha=0.7, fill=False, linewidth=4)

plt.gca().add_patch(circle)
plt.axis('equal')
plt.axis([-1, 1, -1, 1])
plt.title("$\\pi \\approx $" + str(pi_approx))
plt.show()
```
